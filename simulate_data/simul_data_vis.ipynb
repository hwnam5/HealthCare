```python
import pandas as pd

file_path = "squat_data.xlsx"

df = pd.read_excel(file_path)

print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

palette = {0: "blue", 1: "red"}

sns.scatterplot(x=df["UWB"], y=df["accel_x"], hue=df["updown"], palette=palette, ax=axes[0, 0]).set(title="UWB vs Accel X")
sns.scatterplot(x=df["UWB"], y=df["accel_y"], hue=df["updown"], palette=palette, ax=axes[0, 1]).set(title="UWB vs Accel Y")
sns.scatterplot(x=df["UWB"], y=df["accel_z"], hue=df["updown"], palette=palette, ax=axes[0, 2]).set(title="UWB vs Accel Z")
sns.scatterplot(x=df["UWB"], y=df["gyro_x"], hue=df["updown"], palette=palette, ax=axes[1, 0]).set(title="UWB vs Gyro X")
sns.scatterplot(x=df["UWB"], y=df["gyro_y"], hue=df["updown"], palette=palette, ax=axes[1, 1]).set(title="UWB vs Gyro Y")
sns.scatterplot(x=df["UWB"], y=df["gyro_z"], hue=df["updown"], palette=palette, ax=axes[1, 2]).set(title="UWB vs Gyro Z")

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df[["UWB", "accel_x", "accel_y", "accel_z", "gyro_x", "gyro_y", "gyro_z"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Between UWB and Sensor Data")
plt.show()
```

```python
df_40 = df.head(40)

columns_to_plot = ["UWB", "accel_x", "accel_y", "accel_z", "gyro_x", "gyro_y", "gyro_z"]

fig, axes = plt.subplots(len(columns_to_plot), 1, figsize=(10, 15), sharex=True)

for i, col in enumerate(columns_to_plot):
    axes[i].plot(df_40.index, df_40[col], marker="o", linestyle="-", color="orange")
    axes[i].set_ylabel(col)
    axes[i].set_title(f"{col} change")

axes[-1].set_xlabel("Sample Index")

plt.tight_layout()
plt.show()
```
